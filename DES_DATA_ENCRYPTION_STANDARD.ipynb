```python
import binascii

key = input("Enter 8-char key: ").encode()
msg = input("Enter message: ").encode()

# Encrypt (simple XOR)
enc = bytes([m ^ key[i % len(key)] for i, m in enumerate(msg)])
print("Encrypted (hex):", binascii.hexlify(enc).decode())

# Decrypt (XOR again with same key)
dec = bytes([e ^ key[i % len(key)] for i, e in enumerate(enc)])
print("Decrypted:", dec.decode())
```

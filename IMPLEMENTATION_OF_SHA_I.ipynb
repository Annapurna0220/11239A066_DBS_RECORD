```python
import hashlib

def sha1_hash(msg):
    return hashlib.sha1(msg.encode()).hexdigest().upper()

messages = ["", "abc", "abcdefghijklmnopqrstuvwxyz"]

for msg in messages:
    print(f'SHA1("{msg}") = {sha1_hash(msg)}')
```
